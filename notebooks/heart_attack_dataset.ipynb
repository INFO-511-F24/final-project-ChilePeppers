dataset: heart_attack_dataset.csv  
source: https://www.kaggle.com/datasets/waqi786/heart-attack-dataset

The dataset provides gender, age, bloood pressure, cholesterol, diabetes, smoking status, chest pain type, and treament summary for 1000 patients. The dataset has shape 1000 rows 8 columns. The dataset could be used to complete a statistical analysis of heart disease and complete hypothesis testing. The dataset could also be used to train a machine learning algoritm to predict treatment based on patient profile. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/INFO-511-F24/final-project-ChilePeppers/main/data/heart_attack_dataset.csv"

data = pd.read_csv(url)

data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Cholesterol (mg/dL)', hue='Gender', kde=True, stat="density")
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Density')
plt.title('Cholesterol Distribution by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data,x='Blood Pressure (mmHg)',y='Cholesterol (mg/dL)', hue='Treatment')
plt.title('\nHeart Attack Dataset: Blood Pressure vs. Cholesterol\n')
plt.show()
```
