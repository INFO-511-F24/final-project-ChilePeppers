source: https://www.kaggle.com/datasets/patricklford/global-ev-sales-2010-2024


The dataset summarizing the EV market growth in 54 regions/countries during the period 2011 to 2024. The dataset can be used to plot growth by region over time on an interactive map. The data could also be used to plot growth by region, fit a model, and forecast future growth. The dataset shape is 12654 rows and 8 columns. 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#data = pd.read_csv(r'data\IEA Global EV Data 2024.csv')

url = "https://raw.githubusercontent.com/INFO-511-F24/final-project-ChilePeppers/main/data/data/IEA_Global_EV_Data_2024.csv"

data = pd.read_csv(url)

data.head()
```

```python
unique_regions = data['region'].nunique()

print(f'The dataset summarizing {unique_regions} unique regions/countries.')
```

```python
data.info()
```

```python
oz_data = data[data['region'] == 'Australia']
oz_data.head()
```

```python
Oz_data = data[data['region'] == 'Australia']

#Oz_data = Oz_data.groupby(['year', 'parameter'])

Oz_data.head()
```

```python
plt.figure(figsize=(10, 6))
for para in Oz_data['parameter'].unique():
    para_data = Oz_data[Oz_data['parameter'] == para]
    plt.plot(para_data['year'], para_data['value'], label=para)

plt.yscale('log')
plt.legend(title='Parameter') 
plt.xlabel('Year')
plt.ylabel('Value') 
plt.title('Change in Australia EV market from 2011 to 2024')
plt.show()
```
